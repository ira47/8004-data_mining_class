```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

## 1. Incorrect number of clusters

```python
n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.figure(figsize=(12, 5))

plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1])
plt.title('Original Data')

plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Clustering Results")
```

## 2. Anisotropicly distributed data

```python
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.figure(figsize=(12, 5))

plt.subplot(121)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1])
plt.title('Original Data')

plt.subplot(122)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Clustering Results")
```

## 3. Different variance

```python
X_varied, y_varied = make_blobs(n_samples=n_samples,
                                cluster_std=[1.0, 2.5, 0.5],
                                random_state=random_state)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)

plt.figure(figsize=(12, 5))

plt.subplot(121)
plt.scatter(X_varied[:, 0], X_varied[:, 1])
plt.title("Original Data")

plt.subplot(122)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Clustering Results")
```

## 4. Unevenly sized blobs

```python
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = KMeans(n_clusters=3,
                random_state=random_state).fit_predict(X_filtered)

plt.figure(figsize=(12, 5))

plt.subplot(121)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1])
plt.title("Original Data")

plt.subplot(122)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Clustering Results")

plt.show()
```
